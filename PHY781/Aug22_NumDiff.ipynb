## Solving Differential Equations - Numerical method

We shall proceed with solving the equation:
$$ y' = \frac{-y}{x} $$

Using forward difference:

$$ \frac{y(x + h) - y(x)}{h} = \frac{-y(x)}{x} $$
so that
$$ y(x+h) = y(x) - h\frac{y(x)}{x} $$

We shall use the initial valued equation: 
$$y_{n+1} = y_n - h \frac{y_n}{x_n},\;\;\;\;\;(x_0,y_0) = (1,1)\;\;\;\; for \;\; x = 1,...,5$$

```python
import numpy as np
N = 40 #check error for varied values of N. Choose N = 10, 20, 40, 60
x = np.linspace(1,5,N+1)
h = x[1] - x[0]
y = np.zeros(N+1)
# analytic solution below
y_exact = np.zeros(N+1)
error1 = np.zeros(N+1)
# analytic solution above
# Set initial value for y(x0)
y[0] = 1.0
y_exact[0] = 1.0
print(x)
print(h)
print(y)
```

```python
for n in range(N):
    y[n+1] = y[n] - h*y[n]/x[n]
    y_exact[n+1] = 1.0/x[n+1] #we know the exact solution
    error1[n+1] = np.abs(y_exact[n+1] - y[n+1])
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x, y_exact, c = "r")
plt.plot(x, error1, "o")
plt.grid()
```
