## Our first function: The Lagrangian function

```python
#class implementation
def myprod(x, xvalue, i):
    temp = 1.0
    N = len(x)
    for j in range(N):
        if (i != j): temp = temp * (xvalue - x[j])
    return temp

#usage:
    #For Numerator: myprod(x, xvalue, i)
    #For Denominator: myprod(x, x[i], i)
```

```python
x = [0, 1, 2, 3, 4]
f = [1.0, 3.0, 4.0, 2.0, 5.5]
```

```python
xvalue = 2.3
N = len(x)
myfvalue = 0.0
for i in range(N):
    myfvalue += f[i] * myprod(x, xvalue, i)/myprod(x, x[i], i)

print(xvalue, myfvalue)
```

> (a) Calculate the extrapolated valuues of $f(x)$ for xvalues in $[0.0, 4.0]$ in steps of 0.1  
> (b) Plot your results in (a) vs the xvalues  
> (c) Plot the $(x_i, f_i)$ for $i = 1,2...,5$



Solution to (a)

```python
x = [0, 1, 2, 3, 4]
f = [1.0, 3.0, 4.0, 2.0, 5.5]

xvalue = []
fvalue = []

#Get new x values for interpolation
m = 0
for l in range(41):
    xval = round(m + (l * 0.1), 1)
    xvalue.append(xval)

# Interpolate values for each new x values

N = len(x)
def fval(val,N):
    myfvalue = 0.0
    for i in range(N):
        myfvalue += f[i] * myprod(x, val, i)/myprod(x, x[i], i)
    return myfvalue

for val in xvalue:
    fv = fval(val,N)
    fvalue.append(round(fv, 3))

#list of solution
solution = []
for n in range(len(xvalue)):
    solution.append((xvalue[n], fvalue[n]))
```

```python
import matplotlib.pyplot as plt
plt.plot(xvalue,fvalue, marker="o") #this shows line graph
plt.plot(x,f, marker = "+")
plt.legend(["interpolated values", "given values"])
plt.xlabel("x values")
plt.ylabel("f values")
plt.grid() #this adds gridline
plt.title("Plot of interpolated values of x in [0, 4]")
```
